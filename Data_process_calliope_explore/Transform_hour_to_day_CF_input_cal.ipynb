```python
path_inf = '/home/users/train028/euro-calliope-uk_irl/national_rcp/'

hre     = pd.read_csv(path_inf + 'capacityfactors-hydro-reservoir-inflow.csv')
hror    = pd.read_csv(path_inf + 'capacityfactors-hydro-ror.csv')
op_pv   = pd.read_csv(path_inf + 'capacityfactors-open-field-pv.csv')
roof_pv = pd.read_csv(path_inf + 'capacityfactors-rooftop-pv.csv')
wind_of = pd.read_csv('/home/users/train028/euro-calliope-uk_irl/national/capacityfactors-wind-offshore.csv')

elec_demand = pd.read_csv(path_inf + 'electricity-demand.csv')

```

```python
op_pv.columns= ['time','IRL','GBR']
roof_pv.columns= ['time','IRL','GBR']
elec_demand.columns= ['time','IRL','GBR']
```

```python
wind_of_sel = wind_of[['timestep','IRL','GBR']]
wind_of_sel.columns=['time','IRL','GBR']
```

```python
def get_daily(mdat_nat, fname):
    
    mdat_nat['time'] =  pd.to_datetime(mdat_nat['time'], format='%Y-%m-%d %H:%M:%S')
    #mdat.set_index("time").resample("D").ffill().reset_index().rename(
    #{"Date": "Timestamp"}, axis=1)
    day_dat = mdat_nat.resample('d', on='time').mean()
    day_dat.reset_index(level=0, inplace=True)
    day_dat.to_csv(fname + ".csv", sep = ",", index = False)
    
    return(day_dat)
```

```python
get_daily(wind_of_sel,'daily_capacityfactors-wind-offshore')
```

```python
get_daily(hre,'daily_capacityfactors-hydro-reservoir-inflow')
get_daily(hror,'daily_capacityfactors-hydro-ror')
get_daily(op_pv,'daily_capacityfactors-open-field-pv')
get_daily(roof_pv,'daily_capacityfactors-rooftop-pv')
```
